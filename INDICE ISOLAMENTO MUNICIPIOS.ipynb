```python
import pandas as pd
import plotly.express as px 
import re
```

```python
ind_iso_mun = pd.read_csv('ind_iso_ok2.csv', sep=';', encoding='latin-1')
ind_iso_mun
```

```python
ind=ind_iso_mun.drop(['Data'],axis=1)
```

```python
ind_iso_muni=ind.rename(columns={'Município1':'Município', 'Data2':'Data', 'Índice De Isolamento':'Índice De Isolamento (%)'})
#ind_iso_muni

ind_iso=ind_iso_muni.drop(['Índice De Isolamento (%)'], axis=1)
```

```python
#separando os municípios
def split_mun(mun):
    mun = ind_iso.loc[ind_iso['Município']==mun]
    return mun
```

```python
sp=split_mun('SÃO PAULO')
gru=split_mun('GUARULHOS')
camp=split_mun('CAMPINAS')
sb=split_mun('SÃO BERNARDO DO CAMPO')
sjc=split_mun('SÃO JOSÉ DOS CAMPOS')
sa=split_mun('SANTO ANDRÉ')
maua=split_mun('MAUÁ')
soro=split_mun('SOROCABA')
rp=split_mun('RIBEIRÃO PRETO')
osasco=split_mun('OSASCO')
```

```python
fig=px.line(ind_iso, x='Data', y='Índice de isolamento', color='Município',
           labels={'Índice de isolamento': 'Índice de isolamento (%)'},
           #width=800, height=400
           #length=700)
           )

fig.update_layout(yaxis_range=[0,70],
                 title='ÍNDICE DE ISOLAMENTO MUNICIPAL (02/2020 - 10/2021)',
                 )

fig.show()
```
